```
import time

start = time.time()
for index in range(10):
  dt_dot_data = tree.export_graphviz(
    rf_clf.estimators_[index],
    feature_names = X_train.columns,
    class_names = ["Walking", "Walking_Up", "Walking_down", "Sitting", "Standing", "Laying"],
    special_characters=True, # 한글은 특수문자이므로 true로 바꿔줌
    filled = True,
    rounded = True
  )
  dt_graph = pydotplus.graph_from_dot_data(dt_dot_data)
  dt_graph.write_pdf('tree{}.pdf'.format(index))

print('걸린 시간: ', time.time() - start)
```

# RandomForest 성능 향상

```
from sklearn.model_selection import GridSearchCV
```

```
params = {
    'n_estimators':[100], # 트리의 개수 100개
    'max_depth':[6,  8, 10, 12], # 최대 트리의  깊이 6, 8, 10, 12
    'min_samples_leaf':[8, 12, 18], # 최소 트리잎의 개수
    'min_samples_split':[8, 16, 20] # 전체 데이터의 로우 중에서 최소 몇개의 데이터를 이용해서 트리를 만들 것인가
}
```

```
rf_clf = RandomForestClassifier(n_jobs = 10) # CPU를 10개로 분할해서 동시에 10개씩 테스트
```

```
grid_cv = GridSearchCV(rf_clf, param_grid=params, n_jobs = 10)
```

```
import time
start = time.time()
grid_cv.fit(X_train, y_train)
print('걸린시간 : {:.2f}초'.format(time.time() - start))
```

```
print('최적 하이퍼 파라미터 :\n', grid_cv.best_params_)
print('최고 예측 정확도 : {0:.4f}'.format(grid_cv.best_score_))
```

```
rf_clf1 = RandomForestClassifier(n_estimators=100,
                                 max_depth = 12,
                                 min_samples_leaf = 18,
                                 min_samples_split = 20)
rf_clf1
```

```
rf_clf1.fit(X_train,  y_train)
```

```
pred = rf_clf1.predict(X_test)
pred
```

```
print('예측 정확도 : {}'.format(accuracy_score(y_test, pred)))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
ftr_importances_value = rf_clf1.feature_importances_
ftr_importances_value
```

```
ftr_importances = pd.Series(ftr_importances_value, index = X_train.columns)
ftr_importances
```

```
ftr_top20 = ftr_importances.sort_values(ascending=False)[:20]
ftr_top20
```

```
plt.style.use('ggplot')
plt.figure(figsize=(8, 6))
plt.title('Feature importances Top 20')
sns.barplot(x=ftr_top20, y = ftr_top20.index)
plt.show()
```
