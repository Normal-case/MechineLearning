```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# SO2 = 이산화 황, CO = 일산화탄소, O3 = 오존, NO2 = 이산화질소, PM10 = 입자크기 10MM, PM25 = 입자크기 25MM
df = pd.read_csv('2015_data.csv', encoding = 'cp949'); df
```

```
data_seoul = df[df.Location.isin(['서울'])]; data_seoul
```

```
data_seoul.corr()
```

```
import seaborn as sns
%matplotlib inline

sns.set(font_scale = 1.5) # font의 크기를 정하는 것!
f, ax = plt.subplots(figsize = (20, 10))

# seaborn 라이브러리를 이용해 heatmap 표현함
sns_heatmap = sns.heatmap(
    data_seoul.corr(), # 그릴 데이터
    annot = True, # 히트맵 안에 값 표시
    fmt = '.4f', # 몇째자리까지 표현할지
    linewidths = 3.5, # 길 사이의 공백
    cmap='RdBu_r' # 색
)
```
