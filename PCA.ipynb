# PCA 예제

````
import numpy as np
import pandas as pd

data = {
    'x1':[0.2, 0.45, 0.33, 0.54, 0.77],
    'x2' : [5.6, 5.87, 6.37, 7.9, 7.87],
    'x3' : [3.56, 2.4, 1.95, 1.32, 0.98],
    'y' : [100, 70, 60, 45, 30]
}
````

````
df = pd.DataFrame(data); df
````

````
X = df[['x1', 'x2', 'x3']]
````

````
X.mean()
````

````
X.std()
````

````
X = X - X.mean(); X
````

````
X.mean()
````

````
X = X/X.std(); X
````

````
print('평균')
print(X.mean())
print('표준 편차')
print(X.std())
````

````
# 줄 - 1만큼 나눔
X_cov = np.dot(X.T, X) / (5-1); X_cov
````

````
X.cov()
````

````
eigen_value, eigen_vector = np.linalg.eig(X_cov)

print('고유값 :',eigen_value)
print('고유벡터 :', eigen_vector)
````

````
eigen_value[0]
````

````
eigen_vector
````

````
eigen_vector.T[0]
````

````
z1 = X.dot(eigen_vector.T[0]); z1
````

````
z2 = X.dot(eigen_vector.T[1]); z2
````

````
z3 = X.dot(eigen_vector.T[2]); z3
````

````
from sklearn.decomposition import PCA
````

````
pca = PCA(n_components = 3) # 3개의 z1, z2, z3를 계산한다는 의미
````

````
pca.fit(X)
````

````
pca.singular_values_ # 고유값! eigen_value
````

````
pca.components_.T # 1열 : e1, 2열 : e2, 3열 : e3
````

````
z = pca.fit_transform(X); z
````

````
eigen_value[0]
````

````
sum(eigen_value)
````

````
eigen_value[0] / sum(eigen_value) * 100 # z1이 전체에서 91%를 차지! 보통 77% 이상 차지하는 고유값 벡터가 있으면 해당 벡터로 회귀분석 실행
````

````
z1 = X.dot(eigen_vector.T[0]); z1
````

````
from sklearn import linear_model
````

````
linear_regression = linear_model.LinearRegression()
````

````
linear_regression.fit(X = X, y = df['y'])
````

````
linear_regression.score(X = X, y = df['y']) # 전체 X를 다 이용하여 한 회귀분석은 R_square가 크긴하지만 overfitting
````

````
linear_regression.fit(X=pd.DataFrame(z1), y = df['y'])
````

````
linear_regression.score(X=pd.DataFrame(z1), y = df['y']) # z1만을 이용해서 회귀분석. 비록 R_square가 X전체보다는 낮지만 95%이면 상당히 양호 그렇기 때문에 overfitting의 우려도 많이 사라짐
````

# PCA를 이용한 보스턴 집값 예측

````
from sklearn.datasets import load_boston
````

````
boston_dataset = load_boston(); boston_dataset
````

````
boston_dataset.feature_names
````

````
boston = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names); boston
````

````
y = boston_dataset.target; y
````

````
boston.mean()
````

````
X = boston - boston.mean(); X
````

````
np.round(X.mean(), 3) # 거의 0이 나옴
````

````
X.std()
````

````
X = X / X.std(); X
````

````
X.std()
````

````
# 위의 과정을 StandardScaler가 만들어줌
from sklearn.preprocessing import StandardScaler
boston = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names); boston
````

````
# 데이터의 범위를 일정하게 만들어주는 객체 생성
scaler = StandardScaler()
# 각 칼럼의 평균 표준편차 계산
scaler.fit(boston)
# 데이터 - 평균 / 표준편차를 계산해서 각 칼럼의 데이터 범위를 일정하게 맞춰서 X_scaled에 대입
X = scaler.transform(boston)
X
````

````
X = pd.DataFrame(X, columns = boston_dataset.feature_names); X
````

````
from sklearn.decomposition import PCA

pca = PCA(n_components=13)
````

````
# 종속변수 x1 ~ x13가 저장된 DataFrame X에서 z1 ~ z13까지 계산
pca.fit(X)
````

````
print('고유값 :', pca.singular_values_)
print('고유값 벡터:\n', pca.components_.T)
````

````
z = pca.transform(X); z 
````

````
# 고유값 특징이 내림차순이라서 순서대로 77%를 넘는 지점까지 찾으면 됨!
sum(pca.singular_values_)
````

````
pca.singular_values_[0:7].sum() / pca.singular_values_.sum()
````

````
pca.singular_values_[0:8].sum() / pca.singular_values_.sum() # 당첨!
````

````
z.shape
````

````
z.T.shape
````

````
z = z.T[0:8]
````

````
z.shape
````

````
z = z.T
````

````
z.shape
````

````
zDF = pd.DataFrame(z); zDF
````

````
from sklearn.linear_model import LinearRegression
# 선형회귀 객체 생성
linear_regression = LinearRegression()
# 선형회귀 실행
linear_regression.fit(zDF, y)
````

````
linear_regression.score(X = zDF, y = y) # 13개 다썼을 때는 74%이기 때문에 성능차이가 거의 없음! 8개로 하는 것이 overfitting을 예방해줌
````

````
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston 

boston_dataset = load_boston()
````

````
boston = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names); boston
````

````
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(boston)
X = scaler.transform(boston); X
````

````
X = pd.DataFrame(X, columns=boston_dataset.feature_names); X
````

````
from sklearn.decomposition import PCA
pca = PCA(n_components=13)
````

````
pca.fit(X)
````

````
z = pca.transform(X)
````

````
pca.singular_values_[0:8].sum() / pca.singular_values_.sum()
````

````
z.shape
````

````
z[0:8].shape
````

````
type(z)
````

````
z[:,0:8].shape
````

````
z = z[:, 0:8]
````

````
zDF = pd.DataFrame(z); zDF
````

````
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(zDF, y)
````

````
linear_regression.score(X = zDF, y = y)
````
